Partiendo del mejor modelo para la clasificación de las enfermedades obtenido en el notebook anterior, se procede a hacer la función.

Para ello, se seleccionan los 3 síntomas que se repiten con mayor frecuencia, puesto que la posibilidad de que el usuario padezca dichos síntomas será mayor.

Se pregunta al usuario si siente alguno de los síntomas y se rellena una lista con 0 o 1 en función de su respuesta.

Se repite la secuencia hasta conseguir una lista que contenga al menos 6 valores, de los cuales mínimos 3 sean positivos.

Como no se pretende preguntar al usuario 133 veces, se cortará la función en caso de que a la 12ª pregunta siga sin tener ningún síntoma. Como las secuencias son de 3 síntomas, se cortará el bucle en la 4ª iteración.

```python
#Seleccionamos los síntomas que se repiten con mayor frecuencia para pedir al usuario.
#Repetimos la secuencia hasta conseguir una lista de mínimo 6 respuestas a síntomas (1 o 0) y mínimo 3 síntomas positivos.
#Como tampoco queremos estar preguntando al usuario 133 veces, cortamos el bucle en la 4ª iteración

def what_disease():
    
    print("You should see a doctor instead of asking a machine... But I can try and make a diagnosis for you.")

    sleep(3)

    grupo=df.copy()

    # Vamos poniendo en dos listas los síntomas sobre los que se le pregunta al usuario y si los tiene o no.

    sintomas=list()

    valores_sintomas=list()

    iteracion=1

    while iteracion <= 4:

        # Ordenamos los síntomas en función de su frecuencia y vamos filtrando de 3 en 3.

        sint_suma=pd.DataFrame(grupo.sum(0))

        sint_suma.drop("Disease", axis=0, inplace = True)

        sint_suma.drop(sintomas, axis =0, inplace=True)

        sint_comun=list(sint_suma.sort_values(0,ascending=False).index)

        sintom=list()

        if iteracion == 1:

            print(" " * 150)
            print("Please indicate if you have the following symptoms: ")
            print(" " * 150)        
            sintom.append(1 if ((input(f"{sint_comun[0]} ")) in ["Yes", "Ye", "Y", "yes", "ye", "y"]) else 0 )
            sintom.append(1 if ((input(f"{sint_comun[1]} ")) in ["Yes", "Ye", "Y", "yes", "ye", "y"]) else 0 )
            sintom.append(1 if ((input(f"{sint_comun[2]} ")) in ["Yes", "Ye", "Y", "yes", "ye", "y"]) else 0 )

        else:
            sintom.append(1 if ((input(f"{sint_comun[0]} ")) in ["Yes", "Ye", "Y", "yes", "ye", "y"]) else 0 )
            sintom.append(1 if ((input(f"{sint_comun[1]} ")) in ["Yes", "Ye", "Y", "yes", "ye", "y"]) else 0 )
            sintom.append(1 if ((input(f"{sint_comun[2]} ")) in ["Yes", "Ye", "Y", "yes", "ye", "y"]) else 0 )


        for sint, val in zip(sint_comun[0:3], sintom):

            grupo = grupo[(grupo[sint]==val)]


        sintomas.extend(sint_comun[0:3])
        valores_sintomas.extend(sintom)

        if (iteracion == 4) & (sum(valores_sintomas)==0):

            print("I'm just a machine. You should be asking a doctor.")

            break

        elif (sum(valores_sintomas) >= 3) & (len(sintomas)>= 6):

            break

        else:

            iteracion+=1

    #Creamos un DataFrame con los síntomas y si los presenta el usuario 

    usuario=pd.DataFrame(columns=["sintomas", "valores"])
    usuario["sintomas"]=sintomas
    usuario["valores"]=valores_sintomas

    #Creamos un DataFrame con los datos del usuario teniendo en cuenta las columnas del df original para hacer luego la predicción

    enfermedad_usuario=pd.DataFrame(columns=df.columns[1:])
    enfermedad_usuario.loc[0]=0

    for sint, val in zip(sintomas, valores_sintomas):

        enfermedad_usuario[sint]=val 


    #Abrimos el modelo y el LabelEncoder de las enfermedades

    with open(file = "mejor_modelo.pkl", mode = "br") as file:
        mejor_modelo = pickle.load(file)

    with open(file = "Disease_label_encoding.pkl", mode = "br") as file:
        disease_label_encoding = pickle.load(file)

    #Predecimos la enfermedad de nuestro usuario e imprimimos la descripción y todas las indicaciones
    X = enfermedad_usuario

    yhat = mejor_modelo.predict(X)

    enfermedad = disease_label_encoding.inverse_transform(yhat)

    print("*" * 150)
    print(" " * 150)
    print(f"My prediction: {enfermedad[0]}")

    descripcion_enfermedad = descripciones[descripciones["Disease"]==enfermedad[0]]["Description"].values[0]
    print(" " * 150)
    print(descripcion_enfermedad)

    pautas_enfermedad = pd.DataFrame(pautas[pautas["Disease"]==enfermedad[0]]).reset_index()

    pautas_enfermedad.drop("index", axis=1, inplace=True)

    pautas_machine = [pautas_enfermedad[col][0] for col in pautas_enfermedad.columns[1:] if pautas_enfermedad[col][0] != "No symptom"]

    print(" " * 150)
    print("Please take the following precautions: ")
    for p in pautas_machine:
        print(f"- {p}.")
```
